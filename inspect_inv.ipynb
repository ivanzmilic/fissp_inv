```python
import pyana 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
obs = pyana.fzread("cube_to_invert.f0")["data"]
obs.shape
```

```python
fit = pyana.fzread("inverted_spectra.f0")["data"]
fit.shape
```

```python
Iqs = np.mean(obs[:,:,0,:10])
```

```python
obs /= Iqs
fit /= Iqs
```

```python
# Now I want a 2x2 plot to compare observed and fitted for 4 different Stokes components 

i = 2
j = 0

fig, axs = plt.subplots(2, 2, figsize=(10, 8))
axs[0, 0].plot(obs[i, j, 0, :], label='Observed I', color='blue')
axs[0, 0].plot(fit[i, j, 0, :], label='Fitted I', color='orange', linestyle='--')
axs[0, 0].set_title('Stokes I')
axs[0, 0].legend()

axs[0, 1].plot(obs[i, j, 1, :], label='Observed Q', color='blue')
axs[0, 1].plot(fit[i, j, 1, :], label='Fitted Q', color='orange', linestyle='--')
axs[0, 1].set_title('Stokes Q')
axs[0, 1].legend()

axs[1, 0].plot(obs[i, j, 2, :], label='Observed U', color='blue')
axs[1, 0].plot(fit[i, j, 2, :], label='Fitted U', color='orange', linestyle='--')
axs[1, 0].set_title('Stokes U')
axs[1, 0].legend()

axs[1, 1].plot(obs[i, j, 3, :], label='Observed V', color='blue')
axs[1, 1].plot(fit[i, j, 3, :], label='Fitted V', color='orange', linestyle='--')
axs[1, 1].set_title('Stokes V')
axs[1, 1].legend()
```
